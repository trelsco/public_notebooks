```python
import math
import numpy as np
import pandas as pd
%pylab inline
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (16,8)
```

```python
df = pd.read_table("data//BHP_20140530.txt", index_col=2, parse_dates=True)
# df["MID"] = (df["Q.ASK_PRICE"] + df["Q.BID_PRICE"]) / 2.0
df.plot(y=["MID_PRICE"])
```

```python
from collections import deque

class Simplemovingaverage():
    def __init__(self, period):
        assert period == int(period) and period > 0, "Period must be an integer >0"
        self.period = period
        self.stream = deque()
 
    def __call__(self, n):
        stream = self.stream
        stream.append(n)    # appends on the right
        streamlength = len(stream)
        if streamlength > self.period:
            stream.popleft()
            streamlength -= 1
        if streamlength == 0:
            average = 0
        else:
            average = sum( stream ) / streamlength
 
        return average
```

```python
tf = 1000
sma_fast = Simplemovingaverage(tf)
sma_slow = Simplemovingaverage(tf*2)
uppers = []
lowers = []
signals = []

prev_slow = None
prev_fast = None

last_cross_index = 0

x = 0
while x < len(df):
    _sma_fast = sma_fast(df["MID_PRICE"][x])
    _sma_slow = sma_slow(df["MID_PRICE"][x])
    
    if _sma_fast and _sma_slow:
        uppers.append(_sma_fast)
        lowers.append(_sma_slow)
 
    
    if prev_slow and prev_fast:
        if prev_fast > prev_slow and _sma_slow >= _sma_fast:
            if x - last_cross_index >= 350:
                print prev_fast, prev_slow, _sma_fast, _sma_slow
                print "[+] SELL @ {}".format(df["MID_PRICE"][x])
                print "index: {0} {1}".format(last_cross_index, x)
                signals.append(df["T.TIMESTAMP"][x])
                last_cross_index = x
        
        if prev_fast < prev_slow and _sma_fast >= _sma_slow:
            if x - last_cross_index >= 350:
                print prev_fast, prev_slow, _sma_fast, _sma_slow
                print "[+] BUY @ {}".format(df["MID_PRICE"][x])
                print "index: {0} {1}".format(last_cross_index, x)
                signals.append(df["T.TIMESTAMP"][x])
                last_cross_index = x
            
    prev_slow = _sma_slow
    prev_fast = _sma_fast
           
    x+=1

len(uppers), len(lowers), len(df)
```

```python
UPPER = pd.DataFrame(uppers, index=range(len(uppers)))
UPPER.index = df.index

LOWER = pd.DataFrame(lowers, index=range(len(lowers)))
LOWER.index = df.index

# MIDDLE = pd.DataFrame(middles, index=range(len(middles)))
# MIDDLE.index = df.index


result_df = pd.DataFrame(index=df.index)
result_df["MID_PRICE"] = df["MID_PRICE"]
result_df["UPPER"] = UPPER
result_df["LOWER"] = LOWER
# result_df["MIDDLE"] = MIDDLE
plt.figure()

result_df["MID_PRICE"].plot()
result_df["UPPER"].plot(secondary_y=True, color="g")
result_df["LOWER"].plot(secondary_y=True, color="r")
# result_df["MIDDLE"].plot(secondary_y=True, color="orange")

for eachDate in signals:
    plt.axvline(eachDate, color='y')
```
